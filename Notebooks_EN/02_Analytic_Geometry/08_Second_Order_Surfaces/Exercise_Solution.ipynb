```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Task 1: Write the equation of a sphere with center P = (1, 2, 3) and radius r = 3
def plot_sphere(center, radius):
    """Plots a sphere given the center and radius."""
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = radius * np.outer(np.cos(u), np.sin(v)) + center[0]
    y = radius * np.outer(np.sin(u), np.sin(v)) + center[1]
    z = radius * np.outer(np.ones(np.size(u)), np.cos(v)) + center[2]
    return x, y, z

# Define center and radius
center = (1, 2, 3)
radius = 3
x, y, z = plot_sphere(center, radius)

# Plot the sphere
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='b', alpha=0.5)
ax.set_title("Sphere: (x-1)^2 + (y-2)^2 + (z-3)^2 = 9")
plt.show()
```

```python
# Task 3: Intersection of two spheres
# Sphere 1: x^2 + y^2 + z^2 = 1
# Sphere 2: (x-1)^2 + y^2 + z^2 = 1

# Solve x = 1/2 and y^2 + z^2 = 3/4
theta = np.linspace(0, 2 * np.pi, 100)
y_circle = np.sqrt(3 / 4) * np.cos(theta)
z_circle = np.sqrt(3 / 4) * np.sin(theta)
x_circle = np.full_like(theta, 0.5)

# Plot the intersection curve
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_circle, y_circle, z_circle, label="Intersection Curve", color="r")
ax.set_title("Intersection Curve of Two Spheres")
ax.legend()
plt.show()

```

```python
# Task 4: Tangent plane to the paraboloid
# Paraboloid: z = (x-1)^2 + y^2 + 1
# Point P = (1, 0, 1)

def plot_paraboloid():
    """Plots the paraboloid z = (x-1)^2 + y^2 + 1."""
    x = np.linspace(0, 2, 50)
    y = np.linspace(-1, 1, 50)
    x, y = np.meshgrid(x, y)
    z = (x - 1)**2 + y**2 + 1
    return x, y, z

def plot_tangent_plane():
    """Plots the tangent plane z = 1."""
    x = np.linspace(0, 2, 50)
    y = np.linspace(-1, 1, 50)
    x, y = np.meshgrid(x, y)
    z = np.full_like(x, 1)
    return x, y, z

# Generate paraboloid and tangent plane
x_p, y_p, z_p = plot_paraboloid()
x_t, y_t, z_t = plot_tangent_plane()

# Plot paraboloid and tangent plane
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_p, y_p, z_p, alpha=0.5, label="Paraboloid")
ax.plot_surface(x_t, y_t, z_t, color='r', alpha=0.5, label="Tangent Plane")
ax.scatter(1, 0, 1, color="k", label="Point P(1, 0, 1)", s=100)
ax.set_title("Paraboloid and Tangent Plane")
plt.show()
```
