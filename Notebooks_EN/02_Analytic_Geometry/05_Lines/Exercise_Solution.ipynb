```python
import numpy as np
import matplotlib.pyplot as plt

# Coordinates of points A and B
A = (1, 2)
B = (3, 4)

# Calculate slope (m) of the line passing through A and B
m = (B[1] - A[1]) / (B[0] - A[0])

# Line equation: y - y1 = m(x - x1), using point A (1, 2)
# y - 2 = m(x - 1)
# y = m(x - 1) + 2

# Create the line using the slope and point A
x_vals = np.linspace(0, 4, 100)
y_vals = m * (x_vals - A[0]) + A[1]

# Plotting the line and points A and B
plt.figure(figsize=(6, 6))
plt.plot(x_vals, y_vals, label=f'Line through A and B', color='b')
plt.scatter(*A, color='r', label='Point A(1,2)')
plt.scatter(*B, color='g', label='Point B(3,4)')
plt.text(A[0] + 0.1, A[1], 'A(1, 2)', color='r')
plt.text(B[0] + 0.1, B[1], 'B(3, 4)', color='g')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.title("Equation of Line Passing through A(1, 2) and B(3, 4)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
