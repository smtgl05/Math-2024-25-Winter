# Derivatives

## Basic Rules of Differentiation
1. **Power Rule**:
   $$
   \frac{d}{dx} \left( x^n \right) = n x^{n-1}
   $$
2. **Constant Rule**:
   $$
   \frac{d}{dx} (C) = 0
   $$
3. **Constant Multiple Rule**:
   $$
   \frac{d}{dx} \left( C f(x) \right) = C \frac{d}{dx} f(x)
   $$
4. **Sum/Difference Rule**:
   $$
   \frac{d}{dx} \left( f(x) \pm g(x) \right) = f'(x) \pm g'(x)
   $$

---

## Derivatives of Common Functions
1. **Exponential Functions**:
   $$
   \frac{d}{dx} \left( e^x \right) = e^x
   $$
   $$
   \frac{d}{dx} \left( a^x \right) = a^x \ln(a)
   $$
2. **Logarithmic Functions**:
   $$
   \frac{d}{dx} \left( \ln(x) \right) = \frac{1}{x}
   $$
   $$
   \frac{d}{dx} \left( \log_a(x) \right) = \frac{1}{x \ln(a)}
   $$
3. **Trigonometric Functions**:
   $$
   \frac{d}{dx} \left( \sin(x) \right) = \cos(x)
   $$
   $$
   \frac{d}{dx} \left( \cos(x) \right) = -\sin(x)
   $$
   $$
   \frac{d}{dx} \left( \tan(x) \right) = \sec^2(x)
   $$
4. **Inverse Trigonometric Functions**:
   $$
   \frac{d}{dx} \left( \arcsin(x) \right) = \frac{1}{\sqrt{1-x^2}}
   $$
   $$
   \frac{d}{dx} \left( \arctan(x) \right) = \frac{1}{1+x^2}
   $$

---

## Product Rule
$$
\frac{d}{dx} \left( f(x) g(x) \right) = f'(x) g(x) + f(x) g'(x)
$$

## Quotient Rule
$$
\frac{d}{dx} \left( \frac{f(x)}{g(x)} \right) = \frac{f'(x)g(x) - f(x)g'(x)}{g(x)^2}
$$

## Chain Rule
$$
\frac{d}{dx} \left( f(g(x)) \right) = f'(g(x)) g'(x)
$$

---

## Higher-Order Derivatives
1. Second derivative:
   $$
   f''(x) = \frac{d^2 f(x)}{dx^2}
   $$

---

## Special Limits Using L'H么pital's Rule
1. $\displaystyle \lim_{x \to 0} \frac{\sin(x)}{x} = 1$
2. $\displaystyle \lim_{x \to \infty} \frac{\ln(x)}{x} = 0$
3. $\displaystyle \lim_{x \to \infty} \frac{\exp(x)}{x} = \infty$

---

## Derivatives in Physics
1. **Velocity**:
   $$
   V(t) = x'(t)
   $$
2. **Acceleration**:
   $$
   a(t) = V'(t) = x''(t)
   $$

---

## Proving $\frac{d}{dx} (\ln(\sin(x))) = \cot(x)$
Using the chain rule:
$$
\frac{d}{dx} (\ln(\sin(x))) = \frac{1}{\sin(x)} \cdot \cos(x) = \cot(x)
$$

#5th question


---


The position of a particle is given by:  
$$
x(t) = 3t^2 - 6t + 1
$$  
Find the velocity \( V(t) = x'(t) \) and acceleration \( a(t) = V'(t) = x''(t) \) of the particle at \( t = 2 \).

---

**Step 1: Find Velocity**

The velocity \( V(t) \) is the rate of change of position:
$$
V(t) = x'(t) = \frac{d}{dt}(3t^2 - 6t + 1)
$$

Differentiate each term:
- \( \frac{d}{dt}(3t^2) = 6t \)
- \( \frac{d}{dt}(-6t) = -6 \)
- \( \frac{d}{dt}(1) = 0 \)

Combine these results:
$$
V(t) = 6t - 6
$$

---

**Step 2: Find Acceleration**

The acceleration \( a(t) \) is the rate of change of velocity:
$$
a(t) = V'(t) = \frac{d}{dt}(6t - 6)
$$

Differentiate each term:
- \( \frac{d}{dt}(6t) = 6 \)
- \( \frac{d}{dt}(-6) = 0 \)

So, the acceleration is constant:
$$
a(t) = 6
$$

---

**Step 3: Evaluate at \( t = 2 \)**

- For velocity:  
  Substitute \( t = 2 \) into \( V(t) = 6t - 6 \):
  $$
  V(2) = 6(2) - 6 = 12 - 6 = 6
  $$

- For acceleration:  
  Since \( a(t) = 6 \) is constant, at \( t = 2 \):
  $$
  a(2) = 6
  $$



**Final Results**
- **Velocity at \( t = 2 \):** \( V(2) = 6 \)
- **Acceleration at \( t = 2 \):** \( a(2) = 6 \)

**1. Evaluate:**  
$$
\lim_{x \to 0} \frac{\sin(x)}{x}
$$

- **Step 1**: Observe the form of the limit.  
  As \( x \to 0 \), both the numerator (\( \sin(x) \)) and denominator (\( x \)) approach \( 0 \).  
  This is an indeterminate form \( \frac{0}{0} \), so we apply **L'H么pital's Rule**.

- **Step 2**: Differentiate numerator and denominator:  
  $$
  \frac{d}{dx}(\sin(x)) = \cos(x), \quad \frac{d}{dx}(x) = 1
  $$

- **Step 3**: Substitute derivatives into the limit:  
  $$
  \lim_{x \to 0} \frac{\sin(x)}{x} = \lim_{x \to 0} \frac{\cos(x)}{1}
  $$

- **Step 4**: Simplify and evaluate:  
  $$
  \cos(0) = 1
  $$

**Final Answer:**  
$$
\lim_{x \to 0} \frac{\sin(x)}{x} = 1
$$

---

**2. Evaluate:**  
$$
\lim_{x \to \infty} \frac{\ln(x)}{x}
$$

- **Step 1**: Observe the form of the limit.  
  As \( x \to \infty \), both \( \ln(x) \to \infty \) and \( x \to \infty \).  
  This is an indeterminate form \( \frac{\infty}{\infty} \), so we apply **L'H么pital's Rule**.

- **Step 2**: Differentiate numerator and denominator:  
  $$
  \frac{d}{dx}(\ln(x)) = \frac{1}{x}, \quad \frac{d}{dx}(x) = 1
  $$

- **Step 3**: Substitute derivatives into the limit:  
  $$
  \lim_{x \to \infty} \frac{\ln(x)}{x} = \lim_{x \to \infty} \frac{\frac{1}{x}}{1} = \lim_{x \to \infty} \frac{1}{x}
  $$

- **Step 4**: Simplify and evaluate:  
  $$
  \frac{1}{x} \to 0 \quad \text{as } x \to \infty
  $$

**Final Answer:**  
$$
\lim_{x \to \infty} \frac{\ln(x)}{x} = 0
$$

---

**3. Evaluate:**  
$$
\lim_{x \to \infty} \frac{\exp(x)}{x}
$$

- **Step 1**: Observe the form of the limit.  
  As \( x \to \infty \), both \( \exp(x) \to \infty \) and \( x \to \infty \).  
  This is an indeterminate form \( \frac{\infty}{\infty} \), so we apply **L'H么pital's Rule**.

- **Step 2**: Differentiate numerator and denominator:  
  $$
  \frac{d}{dx}(\exp(x)) = \exp(x), \quad \frac{d}{dx}(x) = 1
  $$

- **Step 3**: Substitute derivatives into the limit:  
  $$
  \lim_{x \to \infty} \frac{\exp(x)}{x} = \lim_{x \to \infty} \frac{\exp(x)}{1} = \exp(x)
  $$

- **Step 4**: Simplify and evaluate:  
  Since \( \exp(x) \to \infty \) as \( x \to \infty \), the limit diverges to infinity.

**Final Answer:**  
$$
\lim_{x \to \infty} \frac{\exp(x)}{x} = \infty
$$


##2nd question



Prove that:
$$
\frac{d}{dx} (\ln(\sin(x))) = \cot(x)
$$

### Solution:
1. Start with the derivative of the natural logarithm:
   $$
   \frac{d}{dx} (\ln(\sin(x))) = \frac{1}{\sin(x)} \cdot \frac{d}{dx}(\sin(x))
   $$

2. The derivative of \( \sin(x) \) is \( \cos(x) \):
   $$
   \frac{d}{dx} (\ln(\sin(x))) = \frac{\cos(x)}{\sin(x)}
   $$

3. Simplify \( \frac{\cos(x)}{\sin(x)} \) to \( \cot(x) \):
   $$
   \frac{d}{dx} (\ln(\sin(x))) = \cot(x)
   $$

Thus, the proof is complete.

## 3. For \( f(x) = \cos(x) \), verify that \( f''(x) = -f(x) \).

### Problem Statement:
Verify that:
$$
f''(x) = -f(x)
$$
where \( f(x) = \cos(x) \).

### Solution:
1. Compute the first derivative:
   $$
   f'(x) = \frac{d}{dx} (\cos(x)) = -\sin(x)
   $$

2. Compute the second derivative:
   $$
   f''(x) = \frac{d}{dx} (-\sin(x)) = -\cos(x)
   $$

3. Compare \( f''(x) \) with \( -f(x) \):
   Since \( f(x) = \cos(x) \), we find:
   $$
   f''(x) = -f(x)
   $$

## derivatives of the given functions:

1. **\( y(x) = -3x + 3 \):**
   $$
   y'(x) = -3
   $$

2. **\( y(x) = \pi x + \sin(1) \):**
   - \( \sin(1) \) is a constant, so its derivative is 0.
   $$
   y'(x) = \pi
   $$

3. **\( y(x) = 4 + \sin(2) \):**
   - Both terms are constants, so:
   $$
   y'(x) = 0
   $$

4. **\( y(x) = 2x^3 - 3x^2 + 8x - 9 \):**
   Using the power rule:
   $$
   y'(x) = 6x^2 - 6x + 8
   $$

5. **\( y(x) = 6x^{1/3} \):**
   Using the power rule:
   $$
   y'(x) = 6 \cdot \frac{1}{3} x^{-2/3} = 2x^{-2/3}
   $$

6. **\( y(x) = \sqrt{x} \):**
   Rewrite \( \sqrt{x} \) as \( x^{1/2} \), then use the power rule:
   $$
   y'(x) = \frac{1}{2}x^{-1/2} = \frac{1}{2\sqrt{x}}
   $$

7. **\( y(x) = \cos(x) + \sin(x) \):**
   Using the derivatives of trigonometric functions:
   $$
   y'(x) = -\sin(x) + \cos(x)
   $$

8. **\( y(x) = 2\sin(x)\cos(x) \):**
   Use the product rule:
   $$
   y'(x) = 2\cos^2(x) - 2\sin^2(x)
   $$
   Alternatively, simplify using \( \sin(2x) \):
   $$
   y'(x) = \sin(2x)
   $$

9. **\( y(x) = x\sin(x) \):**
   Use the product rule:
   $$
   y'(x) = 1 \cdot \sin(x) + x \cdot \cos(x) = \sin(x) + x\cos(x)
   $$

10. **\( y(x) = (x+1)(x+1) \):**
    Rewrite as \( y(x) = (x+1)^2 \) and use the chain rule:
    $$
    y'(x) = 2(x+1)
    $$

11. **\( y(x) = \frac{x}{x+1} \):**
    Use the quotient rule:
    $$
    y'(x) = \frac{(1)(x+1) - (x)(1)}{(x+1)^2} = \frac{1}{(x+1)^2}
    $$

12. **\( y(x) = (x+1)\exp(x) \):**
    Use the product rule:
    $$
    y'(x) = (1)(\exp(x)) + (x+1)(\exp(x)) = \exp(x) + (x+1)\exp(x) = (x+2)\exp(x)
    $$

13. **\( y(x) = \sin(x^2) \):**
    Use the chain rule:
    $$
    y'(x) = \cos(x^2) \cdot 2x = 2x\cos(x^2)
    $$

14. **\( y(x) = \exp(-2x) \):**
    Use the chain rule:
    $$
    y'(x) = \exp(-2x) \cdot (-2) = -2\exp(-2x)
    $$

15. **\( y(x) = \frac{1}{\sin(x+1)} \):**
    Rewrite as \( y(x) = (\sin(x+1))^{-1} \) and use the chain rule:
    $$
    y'(x) = -\frac{\cos(x+1)}{\sin^2(x+1)}
    $$

16. **\( y(x) = \sqrt{2x+1} \):**
    Rewrite as \( y(x) = (2x+1)^{1/2} \) and use the chain rule:
    $$
    y'(x) = \frac{1}{2}(2x+1)^{-1/2} \cdot 2 = \frac{1}{\sqrt{2x+1}}
    $$
