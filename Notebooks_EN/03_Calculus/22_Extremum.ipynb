# Extremum Notes: Useful Formulas



## Key Formulas and Steps

### 1. Critical Points

To find critical points of a function $f(x)$:

- Compute the first derivative:
  $$f'(x)$$
- Solve for $x$ where:
  $$f'(x) = 0$$
  or where $f'(x)$ is undefined.

These $x$ values are the critical points.

### 2. First Derivative Test

Determine the nature of a critical point by analyzing the sign changes of $f'(x)$:

- If $f'(x)$ changes from $+$ to $-$, the point is a **local maximum**.
- If $f'(x)$ changes from $-$ to $+$, the point is a **local minimum**.
- If there is no sign change, the point is neither.

### 3. Second Derivative Test

Evaluate the second derivative $f''(x)$ at the critical points:

- If $f''(x) > 0$, the point is a **local minimum**.
- If $f''(x) < 0$, the point is a **local maximum**.
- If $f''(x) = 0$, the test is inconclusive.

### 4. Global Extrema

For a function on a closed interval $[a, b]$, compare:

- Values of $f(x)$ at the critical points within $[a, b]$.
- Values of $f(x)$ at the endpoints $x = a$ and $x = b$.

The largest value is the global maximum, and the smallest value is the global minimum.

## Common Functions and Derivatives

### 1. Power Rule
If $f(x) = x^n$, then:
$$f'(x) = n \cdot x^{n-1}$$
$$f''(x) = n(n-1) \cdot x^{n-2}$$

### 2. Exponential Functions
If $f(x) = e^x$, then:
$$f'(x) = e^x$$
$$f''(x) = e^x$$

If $f(x) = a^x$ (where $a > 0$):
$$f'(x) = a^x \ln(a)$$
$$f''(x) = a^x \ln^2(a)$$

### 3. Logarithmic Functions
If $f(x) = \ln(x)$, then:
$$f'(x) = \frac{1}{x}$$
$$f''(x) = -\frac{1}{x^2}$$

### 4. Trigonometric Functions
- If $f(x) = \sin(x)$:
  $$f'(x) = \cos(x)$$
  $$f''(x) = -\sin(x)$$

- If $f(x) = \cos(x)$:
  $$f'(x) = -\sin(x)$$
  $$f''(x) = -\cos(x)$$

- If $f(x) = \tan(x)$:
  $$f'(x) = \sec^2(x)$$
  $$f''(x) = 2\sec^2(x)\tan(x)$$

### 5. Product Rule
If $f(x) = u(x) \cdot v(x)$:
$$f'(x) = u'(x)v(x) + u(x)v'(x)$$

### 6. Quotient Rule
If $f(x) = \frac{u(x)}{v(x)}$:
$$f'(x) = \frac{u'(x)v(x) - u(x)v'(x)}{v^2(x)}$$

### 7. Chain Rule
If $f(x) = g(h(x))$:
$$f'(x) = g'(h(x)) \cdot h'(x)$$

## Optimization Problems

- Define the function $f(x)$ representing the quantity to be optimized.
- Identify the domain of $f(x)$.
- Find the critical points and endpoints in the domain.
- Use the methods above to determine which points give the desired extrema.

By applying these formulas and steps systematically, you can efficiently solve extremum problems!

\[
P(u) = -2u^2 + 50u - 300
\]

We are tasked with finding the number of units \( u \) that maximize profit.

---

## Step 1: Find the First Derivative
To locate the critical points, compute the first derivative of the profit function:

\[
P'(u) = \frac{d}{du} \big[-2u^2 + 50u - 300\big] = -4u + 50
\]

---

## Step 2: Solve for Critical Points
Set \( P'(u) = 0 \) to find the critical points:

\[
-4u + 50 = 0
\]

Solve for \( u \):

\[
u = \frac{50}{4} = \frac{25}{2}
\]

---

## Step 3: Confirm Maximum or Minimum
To determine if the critical point corresponds to a maximum or minimum, compute the second derivative of \( P(u) \):

\[
P''(u) = \frac{d}{du} \big[-4u + 50\big] = -4
\]

The second derivative is constant and equal to \( -4 \), which is less than \( 0 \). This indicates that the critical point \( u = \frac{25}{2} \) is a **maximum**.

---

## Final Answer:
The number of units that maximize profit is:

\[
u = \frac{25}{2} = 12.5
\]


```python
import sympy as sp

# Define the variable and the profit function
u = sp.Symbol('u')
P = -2 * u**2 + 50 * u - 300

# Find the first derivative of the profit function
dP_du = sp.diff(P, u)

# Solve for the critical points (where the derivative is zero)
critical_points = sp.solve(dP_du, u)

# Check the second derivative to confirm maximum or minimum
second_derivative = sp.diff(dP_du, u)

# Evaluate second derivative at critical points
for point in critical_points:
    test_value = second_derivative.subs(u, point)
    if test_value < 0:
        print(f"The number of units that maximize profit is: {point}")
    else:
        print(f"{point} is not a maximum.")
```

# Problem: Maximizing the Area of a Rectangle
The total perimeter of a rectangle is given as:

\[
\text{Perimeter} = 10 \, \text{m}
\]

We need to find the dimensions \( x \) and \( y \) that maximize the area of the rectangle.

---

## Step 1: Express the Area Function
The perimeter constraint can be written as:

\[
2x + 2y = 10 \quad \Rightarrow \quad x + y = 5 \quad \Rightarrow \quad y = 5 - x
\]

The area of the rectangle is given by:

\[
A(x) = x \cdot y = x \cdot (5 - x) = 5x - x^2
\]

---

## Step 2: Find the Critical Point
To maximize the area, compute the first derivative of \( A(x) \) with respect to \( x \):

\[
\frac{dA}{dx} = \frac{d}{dx} \big[ 5x - x^2 \big] = 5 - 2x
\]

Set the derivative equal to zero to find the critical points:

\[
5 - 2x = 0 \quad \Rightarrow \quad x = \frac{5}{2} = 2.5
\]

---

## Step 3: Verify Maximum Area
Compute the second derivative of \( A(x) \) to confirm whether the critical point corresponds to a maximum:

\[
\frac{d^2A}{dx^2} = \frac{d}{dx} \big[ 5 - 2x \big] = -2
\]

Since the second derivative is negative (\( -2 < 0 \)), the critical point \( x = 2.5 \) corresponds to a maximum.

---

## Step 4: Calculate the Dimensions
Using the perimeter constraint \( y = 5 - x \), the corresponding value of \( y \) is:

\[
y = 5 - 2.5 = 2.5
\]

Thus, the dimensions of the rectangle that maximize the area are:

\[
x = 2.5 \, \text{m}, \quad y = 2.5 \, \text{m}.
\]


```python
# Define the total perimeter
perimeter = 10

# Express the area as a function of one side of the rectangle (x)
def area(x):
    y = (perimeter / 2) - x  # Use the perimeter constraint to find y
    return x * y

# Find the value of x that maximizes the area
import sympy as sp

# Define the variable
x = sp.Symbol('x')

# Define the area function symbolically
y = (perimeter / 2) - x
A = x * y

# Take the derivative of the area function
dA_dx = sp.diff(A, x)

# Solve for the critical point (where the derivative is zero)
critical_point = sp.solve(dA_dx, x)[0]

# Calculate the corresponding y dimension
y_value = (perimeter / 2) - critical_point

# Output the results
print(f"The dimensions of the rectangle that maximize the area are: {critical_point} m x {y_value} m")
```

# Problem: Analyze the Function
The given function is:

\[
f(x) = x^2 + 3x - 5
\]

We are tasked with finding the critical points, determining their nature, and evaluating the function at these points.

---

## Step 1: Compute the First Derivative
The first derivative of \( f(x) \) is:

\[
f'(x) = \frac{d}{dx} \big[x^2 + 3x - 5\big] = 2x + 3
\]

Critical points occur when \( f'(x) = 0 \). Solve:

\[
2x + 3 = 0 \quad \Rightarrow \quad x = -\frac{3}{2}
\]

---

## Step 2: Compute the Second Derivative
The second derivative of \( f(x) \) is:

\[
f''(x) = \frac{d}{dx} \big[2x + 3\big] = 2
\]

The second derivative is constant and positive (\( f''(x) = 2 > 0 \)), which indicates that the critical point \( x = -\frac{3}{2} \) is a **local minimum**.

---

## Step 3: Evaluate the Function at the Critical Point
Substitute \( x = -\frac{3}{2} \) into \( f(x) \) to find the function value:

\[
f\left(-\frac{3}{2}\right) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5
\]

Simplify:

\[
f\left(-\frac{3}{2}\right) = \frac{9}{4} - \frac{9}{2} - 5 = -\frac{29}{4}
\]

---

## Final Answer:
- The critical point is at \( x = -\frac{3}{2} \),
- The function has a **local minimum** at \( x = -\frac{3}{2} \),
- The minimum value of the function is:

\[
f\left(-\frac{3}{2}\right) = -\frac{29}{4}.
\]

---



```python
import sympy as sp

# Define the variable and the function
x = sp.Symbol('x')
f = x**2 + 3*x - 5

# Find the first derivative
df_dx = sp.diff(f, x)

# Solve for critical points
critical_points = sp.solve(df_dx, x)

# Find the second derivative
second_derivative = sp.diff(df_dx, x)

# Evaluate the nature of critical points and the function value
for point in critical_points:
    second_derivative_value = second_derivative.subs(x, point)
    function_value = f.subs(x, point)
    if second_derivative_value > 0:
        print(f"Local minimum at x = {point}, f(x) = {function_value}")
    elif second_derivative_value < 0:
        print(f"Local maximum at x = {point}, f(x) = {function_value}")
    else:
        print(f"Inconclusive at x = {point}, f(x) = {function_value}")
```

\[
f(x) = \frac{x^2 + 2x + 1}{x - 1}
\]

We are tasked with finding the critical points, determining their nature, and evaluating the function at these points.

---

## Step 1: Simplify the Function
The numerator \( x^2 + 2x + 1 \) can be factored as:

\[
x^2 + 2x + 1 = (x + 1)^2
\]

Thus, the function becomes:

\[
f(x) = \frac{(x + 1)^2}{x - 1}
\]

---

## Step 2: Compute the First Derivative
Using the quotient rule for derivatives, where:

\[
f(x) = \frac{g(x)}{h(x)}, \quad f'(x) = \frac{g'(x)h(x) - g(x)h'(x)}{h(x)^2}
\]

Here:
- \( g(x) = (x + 1)^2 \)
- \( h(x) = x - 1 \)

Compute \( g'(x) \) and \( h'(x) \):
\[
g'(x) = 2(x + 1), \quad h'(x) = 1
\]

Apply the quotient rule:
\[
f'(x) = \frac{2(x + 1)(x - 1) - (x + 1)^2}{(x - 1)^2}
\]

Simplify the numerator:
\[
f'(x) = \frac{2(x + 1)(x - 1) - (x + 1)^2}{(x - 1)^2} = \frac{2(x^2 - 1) - (x^2 + 2x + 1)}{(x - 1)^2}
\]

Expand and simplify further:
\[
f'(x) = \frac{2x^2 - 2 - x^2 - 2x - 1}{(x - 1)^2} = \frac{x^2 - 2x - 3}{(x - 1)^2}
\]

Factorize the numerator:
\[
f'(x) = \frac{(x - 3)(x + 1)}{(x - 1)^2}
\]

---

## Step 3: Solve for Critical Points
Critical points occur when \( f'(x) = 0 \):

\[
(x - 3)(x + 1) = 0 \quad \Rightarrow \quad x = 3 \, \text{or} \, x = -1
\]

Note: \( x = 1 \) is excluded because it makes the denominator zero.

Compute the second derivative \( f''(x) \) or analyze the sign of \( f'(x) \) around the critical points.

## Final Answer:
