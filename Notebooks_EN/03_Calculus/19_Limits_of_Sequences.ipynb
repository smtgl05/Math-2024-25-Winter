# Limits of Sequences

## Useful Formulas for Solving Limit Problems

### **1. General Limit Properties**
- **Polynomial Limits (as n to infty):**
  $$
  \lim_{n \to \infty} \frac{P(n)}{Q(n)} =
  \begin{cases}
  0, & \text{if the degree of } P(n) < Q(n), \\
  \infty \text{ or } -\infty, & \text{if the degree of } P(n) > Q(n), \\
  \frac{\text{leading coefficient of } P(n)}{\text{leading coefficient of } Q(n)}, & \text{if the degree of } P(n) = Q(n).
  \end{cases}
  $$

- **Behavior of Fractions:**
  $$
  \frac{1}{n^k} \to 0, \quad \text{as } n \to \infty \quad (k > 0).
  $$

---

### **2. Squeeze Theorem**
- If \( a_n \leq b_n \leq c_n \) for all \( n \), and:
  $$
  \lim_{n \to \infty} a_n = \lim_{n \to \infty} c_n = L,
  $$
  then:
  $$
  \lim_{n \to \infty} b_n = L.
  $$

---

### **3. Trigonometric Limits**
- **Basic Limit:**
  $$
  \lim_{n \to \infty} \frac{\sin(n)}{n} = 0
  $$
  - **Reason:** Using the Squeeze Theorem:
    $$
    -\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}.
    $$

---

### **4. Exponential Approximation**
- **Exponential Sequence Limit:**
  $$
  \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = e
  $$

---

### **5. Growth Rates of Sequences**
- For \( n \to \infty \):
  $$
  n^k \gg \ln(n) \quad \text{and} \quad e^n \gg n^k, \quad (k > 0).
  $$

### Find the limit of the sequence:

$$
a_n = \left(1 + \frac{1}{n}\right)^n
$$

---

### **Step-by-Step Solution**

#### Step 1: Recall the definition of \(e\)
From calculus, the constant \(e\) is defined as:

$$
e = \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n
$$

---

#### Step 2: Apply the definition
The given sequence:

$$
a_n = \left(1 + \frac{1}{n}\right)^n
$$

is identical to the expression for \(e\).

---

#### Step 3: Conclude the limit
As \(n \to \infty\):

$$
\lim_{n \to \infty} a_n = e
$$

---

### **Final Answer:**

$$
\lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = e
$$

The limit of the sequence is the constant \(e \approx 2.718\).

#2nd question

### Prove using the Squeeze Theorem:
$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0
$$

---

#### Step 1: The properties of \(\sin(n)\)
- The sine function is bounded:
$$
-1 \leq \sin(n) \leq 1, \quad \text{for all } n.
$$

---

#### Step 2: Divide the inequality by \(n\)
- Dividing \( -1 \leq \sin(n) \leq 1 \) by \(n\) (for \(n > 0\)):
$$
-\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}.
$$

---

#### Step 3: Take the limit of the bounds
- As \(n \to \infty\):
$$
\lim_{n \to \infty} -\frac{1}{n} = 0 \quad \text{and} \quad \lim_{n \to \infty} \frac{1}{n} = 0.
$$

---

#### Step 4: Apply the Squeeze Theorem
- By the Squeeze Theorem:
$$
-\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n},
$$
and since the limits of the lower and upper bounds are both \(0\):
$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0.
$$

---

### Final Answer:
$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0
$$

# First Question

### 1. Calculate:

#### (a)
$$
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n}
$$

**Solution:**
1. Factor out the highest power of \( n \) from both the numerator and the denominator:
   $$
   \frac{n^2 + 3n}{2n^2 - 2n} = \frac{n^2(1 + \frac{3}{n})}{n^2(2 - \frac{2}{n})}
   $$

2. Simplify by canceling \( n^2 \):
   $$
   \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}}
   $$

3. Take the limit as \( n \to \infty \). Since \( \frac{1}{n} \to 0 \):
   $$
   \frac{1 + 0}{2 - 0} = \frac{1}{2}.
   $$

**Final Answer:**
$$
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n} = \frac{1}{2}.
$$

---

#### (b)
$$
\lim_{n \to \infty} \frac{(2n + 3)^3}{n^3 - 1}
$$

**Solution:**
1. Expand \( (2n + 3)^3 \) using the binomial theorem:
   $$
   (2n + 3)^3 = 8n^3 + 12n^2 \cdot 3 + 6n \cdot 9 + 27 = 8n^3 + 36n^2 + 54n + 27.
   $$

2. Rewrite the limit:
   $$
   \frac{(2n + 3)^3}{n^3 - 1} = \frac{8n^3 + 36n^2 + 54n + 27}{n^3 - 1}.
   $$

3. Divide the numerator and denominator by \( n^3 \):
   $$
   \frac{8 + \frac{36}{n} + \frac{54}{n^2} + \frac{27}{n^3}}{1 - \frac{1}{n^3}}.
   $$

4. Take the limit as \( n \to \infty \). Terms involving \( \frac{1}{n^k} \) approach 0:
   $$
   \frac{8 + 0 + 0 + 0}{1 - 0} = 8.
   $$

**Final Answer:**
$$
\lim_{n \to \infty} \frac{(2n + 3)^3}{n^3 - 1} = 8.
$$
