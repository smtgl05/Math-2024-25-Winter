$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$

## **1. Solve the following first-order ordinary differential equations**

### **(a) $ y'(x) = y(x) $**

1. Rewrite the equation:
   $$
   \frac{dy}{dx} = y
   $$

2. Separate the variables:
   $$
   \frac{1}{y} \, dy = dx
   $$

3. Integrate both sides:
   $$
   \int \frac{1}{y} \, dy = \int dx
   $$

   Using the rules:
   $$
   \int \frac{1}{y} \, dy = \ln|y|, \quad \int dx = x + C
   $$

   So:
   $$
   \ln|y| = x + C
   $$

4. Exponentiate both sides:
   $$
   y = e^{x+C} = e^C \cdot e^x
   $$

   Let $ e^C = C_1 $, so:
   $$
   y = C_1 e^x
   $$

---

### **(b) $ y'(x) = \frac{1}{2y(x)} $**

1. Rewrite the equation:
   $$
   \frac{dy}{dx} = \frac{1}{2y}
   $$

2. Separate the variables:
   $$
   2y \, dy = dx
   $$

3. Integrate both sides:
   $$
   \int 2y \, dy = \int dx
   $$

   Using the rules:
   $$
   \int y \, dy = \frac{y^2}{2}, \quad \int dx = x + C
   $$

   So:
   $$
   y^2 = x + C
   $$

4. Solve for $ y $:
   $$
   y = \pm \sqrt{x + C}
   $$

---

## **2. Solve the first-order ordinary differential equations using the method of separation of variables**

### **(a) $ \frac{dy}{dx} = \frac{x}{y} $**

1. Rewrite the equation:
   $$
   y \, dy = x \, dx
   $$

2. Integrate both sides:
   $$
   \int y \, dy = \int x \, dx
   $$

   Using the rules:
   $$
   \int y \, dy = \frac{y^2}{2}, \quad \int x \, dx = \frac{x^2}{2}
   $$

   So:
   $$
   \frac{y^2}{2} = \frac{x^2}{2} + C
   $$

3. Simplify:
   $$
   y^2 = x^2 + 2C
   $$

   Let $ 2C = C_1 $:
   $$
   y^2 = x^2 + C_1
   $$

4. Solve for $ y $:
   $$
   y = \pm \sqrt{x^2 + C_1}
   $$

---

### **(b) $ \frac{dy}{dx} = \frac{y}{x} $**

1. Rewrite the equation:
   $$
   \frac{1}{y} \, dy = \frac{1}{x} \, dx
   $$

2. Integrate both sides:
   $$
   \int \frac{1}{y} \, dy = \int \frac{1}{x} \, dx
   $$

   Using the rules:
   $$
   \int \frac{1}{y} \, dy = \ln|y|, \quad \int \frac{1}{x} \, dx = \ln|x| + C
   $$

   So:
   $$
   \ln|y| = \ln|x| + C
   $$

3. Exponentiate both sides:
   $$
   y = e^{\ln|x| + C} = e^C \cdot |x|
   $$

   Let $ e^C = C_1 $:
   $$
   y = C_1 |x|
   $$

---

## **3. Solve the second-order ordinary differential equations**

### **(a) $ y''(x) + y'(x) = 0 $, with boundary conditions $ y(0) = 2 $ and $y'(0) = -1 $**

1. Assume a solution of the form:
   $$
   y(x) = e^{rx}
   $$

2. Substitute into the equation to get the characteristic equation:
   $$
   r^2 + r = 0
   $$

3. Solve the characteristic equation:
   $$
   r(r + 1) = 0 \quad \implies \quad r = 0 \, \text{or} \, r = -1
   $$

4. General solution:
   $$
   y(x) = C_1 + C_2 e^{-x}
   $$
  

5. Apply boundary conditions:
   - At $ x = 0 $, $ y(0) = 2 $:

     $$
     C_1 + C_2 = 2
     $$

   - At $ x = 0 $, $ y'(0) = 0 $:
   
     $$
     y'(x) = C_1 e^x - C_2 e^{-x}, \quad y'(0) = C_1 - C_2 = 0 \implies C_1 = C_2
     $$


6. Solve for $ C_1 $:
   $$
   C_1 + 1 = 2 \quad \implies \quad C_1 = 1
   $$

7. Final solution:
   $$
   y(x) = 1 + e^{-x}
   $$

---

### **(b) $ y''(x) - y(x) = 0 $, with boundary conditions $ y(0) = 2 $ and $ y'(0) = 0 $**

1. Assume a solution of the form:
   $$
   y(x) = e^{rx}
   $$

2. Substitute into the equation to get the characteristic equation:
   $$
   r^2 - 1 = 0
   $$

3. Solve the characteristic equation:
   $$
   r = 1 \, \text{or} \, r = -1
   $$

4. General solution:
   $$
   y(x) = C_1 e^x + C_2 e^{-x}
   $$

5. Apply boundary conditions:
   - At $ x = 0 $, $ y(0) = 2 $:

     $$
     C_1 + C_2 = 2
     $$
   - At $ x = 0 $, $ y'(0) = 0 $:

     $$
     y'(x) = C_1 e^x - C_2 e^{-x}, \quad y'(0) = C_1 - C_2 = 0 \implies C_1 = C_2
     $$

6. Solve for $ C_1 $ and $ C_2 $:
   $$
   2C_1 = 2 \quad \implies \quad C_1 = 1, \, C_2 = 1
   $$

7. Final solution:
   $$
   y(x) = e^x + e^{-x}
   $$

## **4. Verify if $ \psi(t, x) = A \cos(\omega t + kx) $ solves the wave equation**

The wave equation is:
$$
\frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0, \quad v = \frac{\omega}{k}
$$

1. Compute $ \frac{\partial^2 \psi}{\partial t^2} $:
   $$
   \frac{\partial \psi}{\partial t} = -A \omega \sin(\omega t + kx)
   $$
   $$
   \frac{\partial^2 \psi}{\partial t^2} = -A \omega^2 \cos(\omega t + kx)
   $$

2. Compute $ \frac{\partial^2 \psi}{\partial x^2} $:
   $$
   \frac{\partial \psi}{\partial x} = -A k \sin(\omega t + kx)
   $$
   $$
   \frac{\partial^2 \psi}{\partial x^2} = -A k^2 \cos(\omega t + kx)
   $$

3. Substitute into the wave equation:
   $$
   -A \omega^2 \cos(\omega t + kx) - v^2 (-A k^2 \cos(\omega t + kx)) = 0
   $$

   Using $ v^2 = \frac{\omega^2}{k^2} $:
   $$
   -A \omega^2 \cos(\omega t + kx) + A \omega^2 \cos(\omega t + kx) = 0
   $$

4. Conclusion:
   $$
   \psi(t, x) = A \cos(\omega t + kx) \, \text{satisfies the wave equation.}
   $$
